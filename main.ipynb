```python
# Реализовать модель сети Джордана-Элмана с сигмоидной функцией активации.
import numpy as np
import math
import matplotlib.pyplot as plt
import time
```

```python
def normalize_m(matrix):
    for j in range(matrix.shape[1]):
        s = 0
        for i in range(matrix.shape[0]):
            s += pow(matrix.item((i,j)), 2)
        s = math.sqrt(s)
        
        if s > 1000:
            print(s)

        for i in range(matrix.shape[0]):
            matrix.itemset((i, j), matrix.item((i, j)) / s)
    return matrix
```

```python
def forward(input_nums, x, y, w1_1, w1_2, w1_3, w2):    
    x = F(S(input_nums, x, y, w1_1, w1_2, w1_3))
    y = F(S_(x, w2))
    return x, y
```

```python
def transform(numset, p, m):
    mat = [[0]*p for i in range(m)]
    for i in range(m):
        mat[i] = numset[i: i + p]
        
    return mat
```

```python
def update_w1(w, p, m, alpha, error, x):
    if type(x) is not list:
        x = x.transpose()
    
    for i in range(p):
        for j in range(m):
            w[i][j] -= alpha * error * x[i]
    return w

    
def update_w2(w, m, alpha, error, x):    
    for i in range(m):
        w[i][0] -= alpha * error * x.item(i)
    return w
```
