```python
# Реализовать модель сети Джордана-Элмана с сигмоидной функцией активации.
import numpy as np
import math
import matplotlib.pyplot as plt
import time
```

```python
def F(S):
    result = 1 / (1 + np.exp(-1*S))
#     result = S
    return result

def S(input_nums, x, y, w1_1, w1_2, w1_3, T1):
    layer1_1 = np.matrix(input_nums).reshape(1, p)
    layer1_2 = x
    layer1_3 = y
    
    s = 0
    s += np.matmul(layer1_1, w1_1)
    s += np.matmul(layer1_2, w1_2)
    s += np.matmul(layer1_3, w1_3)
    return s - T1

def S_(x, w2, T2):
    s = np.matmul(x, w2)
    return s - T2
```

```python
def forward(input_nums, x, y, w1_1, w1_2, w1_3, w2, T1, T2):    
    x = F(S(input_nums, x, y, w1_1, w1_2, w1_3, T1))
    y = F(S_(x, w2, T2))
    return x, y
```

```python
def transform(numset, p, m):
    mat = [[0]*p for i in range(m)]
    for i in range(m):
        mat[i] = numset[i: i + p]
        
    return mat
```

```python
def update_w1(w, p, m, alpha, error, x, y):
    if type(x) is not list:
        x = x.transpose()
    y = y.transpose()
    for i in range(p):
        for k in range(m):
            w[i][k] -= alpha * error * x[i]
#             w[i][k] -= alpha * error * x[i] * y[k]
    return w

    
def update_w2(w, m, alpha, error, x, y):    
    for i in range(m):
        w[i][0] -= alpha * error * x.item(i)
    return w
```

```python
def learning(numset, p, m, alpha, e, N, Z):
    """Run learning algoritm"""
    mat = transform(numset, p, m)
    
    w2 = np.random.rand(m, 1) * 2 - 1
    
    w1_1 = np.random.rand(p , m) * 2 - 1
    w1_2 = np.random.rand(m , m) * 2 - 1
    w1_3 = np.random.rand(1 , m) * 2 - 1
        
    x = np.zeros(m).reshape(1, m)
    y = np.zeros(1).reshape(1, 1)
    
    step = 0
    err_max = 0
    time1 = time.time()
        
    while True:
        error_all = 0
        step += 1
        temp = 0
        T1 = 0
        T2 = 0
        for row in mat:
            p_x, p_y = x, y
            x, y = forward(row, x, y, w1_1, w1_2, w1_3, w2, T1, T2)
            
            error = y - numset[p + temp]
            
            w1_1 = update_w1(w1_1, p, m, alpha, error, row, x)
            w1_2 = update_w1(w1_2, m, m, alpha, error, p_x, x)
            w1_3 = update_w1(w1_3, 1, m, alpha, error, p_y, x)
            w2 = update_w2(w2, m, alpha, error, x, y)
            T2 += alpha * error
            T2 += T1
                                  
            error_all += error
            temp += 1
            
#         print(error_all)
        error_all = pow(error_all, 2) / 2
        if step % 10 is 0:
            time2 = time.time()
            print("***step time is ", time2 - time1)
            print(step, " ", error_all)
            time1 = time2
        
        if error_all < e and step > N:
            
            x = np.zeros(m).reshape(1, m)
            y = np.zeros(1).reshape(1, 1)
            mat = transform([1/100, 3/100, 9/100, 27/100, 81/100, 243/100, 729/100], 3, 4)
            for row in mat:
                x, y = forward(row, x, y, w1_1, w1_2, w1_3, w2, T1, T2)
                print(y)
            
            
            return step
    print("tyty")
```
