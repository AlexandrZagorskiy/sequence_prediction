```python
# Реализовать модель сети Джордана-Элмана с сигмоидной функцией активации.
import numpy as np
import math
import matplotlib.pyplot as plt
import time
```

```python
def F(S):
    result = 1 / (1 + np.exp(-1*S))
#     result = S
    return result

def S(input_nums, x, y, w1_1, w1_2, w1_3, T):
    layer1_1 = np.matrix(input_nums).reshape(1, p)
    layer1_2 = x
    layer1_3 = y
    
    s = 0
    s += np.matmul(layer1_1, w1_1)
    s += np.matmul(layer1_2, w1_2)
    s += np.matmul(layer1_3, w1_3)
    return s - T

def S_(x, w2, T):
    s = np.matmul(x, w2)
    return s - T
```

```python
def forward(input_nums, x, y, w1_1, w1_2, w1_3, w2, T):    
    x = F(S(input_nums, x, y, w1_1, w1_2, w1_3, T))
    y = F(S_(x, w2, T))
    return x, y
```

```python
def transform(numset, p, m):
    mat = [[0]*p for i in range(m)]
    for i in range(m):
        mat[i] = numset[i: i + p]
        
    return mat
```

```python
def update_w1(w, p, m, alpha, error, x, y):
    if type(x) is not list:
        x = x.transpose()
    y = y.transpose()
    for i in range(p):
        for k in range(m):
            w[i][k] -= alpha * error * F(x[i]) * (1 - F(x[i])) * y[k]
#             w[i][k] -= alpha * error * x[i] * y[k]
    return w

    
def update_w2(w, m, alpha, error, x, y):    
    for i in range(m):
        w[i][0] -= alpha * error * x.item(i) * y.item(0)
#         w[i][0] -= alpha * error * x.item(i) * (1 - x.item(i)) * y.item(0)
    return w
```
